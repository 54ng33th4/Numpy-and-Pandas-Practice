#Q1. Employee Salary Adjustment

A company stores salaries in a NumPy array.

```python
salaries = np.array([45000, 52000, 61000, 48000, 75000])
print("Original Salaries:", salaries)
```

Tasks:
1. Increase all salaries by 10%.

```python
updated_salaries = salaries * 1.10
print("Updated Salaries (10% increase):", updated_salaries)
```

2. Find the average salary.

```python
average_salary = np.mean(updated_salaries)
print("Average Salary:", average_salary)
```

3. Find the maximum and minimum salary.

```python
max_salary = np.max(updated_salaries)
min_salary = np.min(updated_salaries)

print("Maximum Salary:", max_salary)
print("Minimum Salary:", min_salary)
```

4. Print salaries greater than 60000.

```python
high_salaries = updated_salaries[updated_salaries > 60000]
print("Salaries greater than 60000:", high_salaries)
```

#Q2. Sales Data Analysis

Daily sales of a store for 10 days:

```python
sales = np.array([120,150,180,200,170,160,210,220,190,175])
print("Daily Sales:", sales)
```

Tasks:
1. Compute total sales.

```python
total_sales = np.sum(sales)
print("Total Sales:", total_sales)
```

2. Compute mean and standard deviation.

```python
mean_sales = np.mean(sales)
std_sales = np.std(sales)

print("Mean Sales:", mean_sales)
print("Standard Deviation:", std_sales)
```

3. Identify days where sales exceeded the average.

```python
above_avg_days = np.where(sales > mean_sales)[0] + 1  # +1 for day numbering
above_avg_sales = sales[sales > mean_sales]

print("Days where sales exceeded average:", above_avg_days)
print("Sales on those days:", above_avg_sales)
```

4. Normalize the sales using Min-Max scaling.

```python
min_sales = np.min(sales)
max_sales = np.max(sales)

normalized_sales = (sales - min_sales) / (max_sales - min_sales)

print("Normalized Sales (Min-Max Scaling):", normalized_sales)
```

#Q3. Matrix Operations for Business Data

A company has sales data for 3 products across 4 regions.

```python
# Given data
sales_matrix = np.array([
    [120, 140, 130, 150],
    [80,  90,  100, 95],
    [200, 210, 190, 205]
])

print("Sales Matrix:\n", sales_matrix)
```

Tasks:
1. Find total sales per product.

```python
total_per_product = np.sum(sales_matrix, axis=1)
print("Total Sales per Product:", total_per_product)
```

2. Find total sales per region.

```python
total_per_region = np.sum(sales_matrix, axis=0)
print("Total Sales per Region:", total_per_region)
```

3. Identify the best performing product.

```python
best_product_index = np.argmax(total_per_product)
print("Best Performing Product (Product Index):", best_product_index)
print("Sales of Best Product:", total_per_product[best_product_index])
```

4. Identify the region with highest sales.

```python
best_region_index = np.argmax(total_per_region)
print("Best Performing Region (Region Index):", best_region_index)
print("Sales of Best Region:", total_per_region[best_region_index])
```

#Q5. Missing Data Handling

Dataset:

```python
# Given data
data = {
    'Name': ['John','Anna','Peter','Linda','David'],
    'Age': [28, np.nan, 35, 32, np.nan],
    'Salary': [50000,54000,np.nan,58000,60000]
}

# Create DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

Tasks:
1. Detect missing values.

```python
print("Missing Values (True indicates missing):\n", df.isnull())

print("Missing Values Count:\n", df.isnull().sum())
```

2. Replace missing Age with mean age.

```python
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

print("After Filling Missing Age with Mean:\n", df)
```

3. Replace missing Salary with median salary.

```python
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

print("After Filling Missing Salary with Median:\n", df)
```

4. Remove rows with any remaining missing values.

```python
df_cleaned = df.dropna()

print("Final Cleaned DataFrame:\n", df_cleaned)
```

#Q7. Customer Purchase Analysis

Dataset:

```python
# Given data
data = {
    'CustomerID': [101,102,103,104,105,106,107],
    'Age': [25,45,35,50,23,40,60],
    'PurchaseAmount': [250,600,400,800,150,700,900],
    'City': ['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

# Create DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

Tasks:
1. Find the average purchase amount by city.

```python
avg_purchase_city = df.groupby('City')['PurchaseAmount'].mean()
print("Average Purchase by City:\n", avg_purchase_city)
```

2. Identify customers with purchase above overall average.

```python
overall_avg = df['PurchaseAmount'].mean()
above_avg_customers = df[df['PurchaseAmount'] > overall_avg]

print("Overall Average Purchase:", overall_avg)
print("Customers with Above Average Purchase:\n", above_avg_customers)
```

3. Find the city generating the highest revenue.

```python
total_revenue_city = df.groupby('City')['PurchaseAmount'].sum()
top_city = total_revenue_city.idxmax()

print("Total Revenue by City:\n", total_revenue_city)
print("City with Highest Revenue:", top_city)
```

4. Create a column Age_Group:

*   <30 → Young
*   30–50 → Adult
*   50 → Senior

```python
def categorize_age(age):
    if age < 30:
        return 'Young'
    elif age <= 50:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(categorize_age)
print("DataFrame with Age_Group:\n", df)
```

5. Find average purchase per age group.

```python
avg_purchase_age_group = df.groupby('Age_Group')['PurchaseAmount'].mean()
print("Average Purchase per Age Group:\n", avg_purchase_age_group)
```

#Q8. Time Series Sales Analysis
Generate data:

```python
# Generate data
dates = pd.date_range('2024-01-01', periods=12, freq='ME')  #collab suggested that 'M' is deprecated and will be removed in a future version, use 'ME' instead.
sales = [200,220,250,270,300,320,350,360,380,400,420,450]

df = pd.DataFrame({'Date': dates, 'Sales': sales})
print("Original DataFrame:\n", df)
```

Tasks:
1. Convert Date to datetime index.

```python
df.set_index('Date', inplace=True)
print("DataFrame with Date as Index:\n", df)
```

2. Compute monthly percentage growth.

```python
df['Monthly_Growth (%)'] = df['Sales'].pct_change() * 100
print("Monthly Percentage Growth:\n", df)
```

3. Calculate 3-month rolling average sales.

```python
df['Rolling_Avg_3'] = df['Sales'].rolling(window=3).mean()
print("With 3-Month Rolling Average:\n", df)
```

4. Identify the month with highest growth rate.

```python
max_growth_month = df['Monthly_Growth (%)'].idxmax()
max_growth_value = df['Monthly_Growth (%)'].max()

print("Month with Highest Growth:", max_growth_month)
print("Highest Growth Rate (%):", max_growth_value)
```
